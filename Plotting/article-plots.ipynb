```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd 

plt.rcParams['font.family'] = 'Arial'

data = pd.read_csv('../Pseudo-Labeling/fix_balanced_aaf_evaluation_results.csv')

initial_row = 2 - 2
number_iterations = 10
iterations = np.arange(1, number_iterations+1)
end_row = initial_row + number_iterations

general_acc = data['Accuracy'][initial_row:end_row].to_numpy()
female_acc = data['Female Accuracy'][initial_row:end_row].to_numpy()
male_acc = data['Male Accuracy'][initial_row:end_row].to_numpy()

baseline = 73.28
male_baseline = 97.94
female_baseline = 48.61
difference = np.abs(np.array(male_acc) - np.array(female_acc))
best_iteration = np.argmin(difference) + 1
best_general_acc = np.argmax(general_acc) + 1


plt.figure(figsize=(10, 6))

# Plot accuracy lines
plt.plot(iterations, general_acc, label='General', marker='o', color='#1f77b4', linewidth=2)
plt.plot(iterations, male_acc, label='Male', marker='s', color='#ff7f0e', linewidth=2)
plt.plot(iterations, female_acc, label='Female', marker='d', color='#9467bd', linewidth=2)

# Plot baselines
plt.axhline(y=baseline, color='#7f7f7f', linestyle='--', linewidth=1.5, label='Baseline')
plt.axhline(y=male_baseline, color='#ffbb78', linestyle='--', linewidth=1.5, label='Male Baseline')
plt.axhline(y=female_baseline, color='#c5b0d5', linestyle='--', linewidth=1.5, label='Female Baseline')

# Plot vertical lines for best iteration and lowest gender gap
plt.axvline(x=best_general_acc, color='#17becf', linestyle='--', linewidth=1.5, label='Best General Accuracy')
plt.axvline(x=best_iteration, color='#e377c2', linestyle='--', linewidth=1.5, label='Lowest Gender Gap')

# Labels and title
plt.xlabel('Iterations', fontsize=12)
plt.ylabel('AAF Test Set Accuracy [%]', fontsize=12)

# Legend and grid
plt.legend(loc='upper center', bbox_to_anchor=(0.5, 1.15), ncol=4, fontsize=10)
plt.grid(True, linestyle='--', alpha=0.6)

# Customize ticks
plt.xticks(iterations, fontsize=10)
plt.yticks(fontsize=10)

# Save and show
plt.tight_layout()
plt.savefig('Balanced_FixMatch_0.6_Final.png', dpi=300, bbox_inches='tight')
plt.show()
```
